```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
targets = df['Class']
```

```python
dataset = df.iloc[:, :-1]
```

```python
dataset.head()
```

```python
X = dataset
y = targets
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data, train_targets, test_targets = train_test_split(X, y, test_size=0.3)
print(len(train_data))
```

```python
mean = train_data.mean(axis=0)
train_data -= mean
std = train_data.std(axis=0)
train_data /= std
test_data -= mean
test_data /= std
```

```python
train_data.head()
```

```python
train_data, validation_data, train_targets, validation_targets = train_test_split(train_data, train_targets, test_size=0.287)
```

```python
print(len(train_data))
print(len(test_data))
print(len(validation_data))
```

```python
from keras import models
from keras import layers
from keras import regularizers
```

```python
model1 = models.Sequential()

model1.add(layers.Dense(30, activation='relu', input_shape=(train_data.shape[1],)))

model1.add(layers.Dense(10, activation='relu'))
model1.add(layers.Dropout(0.5))
model1.add(layers.Dense(8, activation='relu'))
model1.add(layers.Dropout(0.5))
model1.add(layers.Dense(6, activation='relu'))
model1.add(layers.Dropout(0.5))

model1.add(layers.Dense(1, activation='sigmoid'))

model1.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

model1.fit(train_data, train_targets, epochs=100, batch_size=128, validation_data=(validation_data, validation_targets))
```
